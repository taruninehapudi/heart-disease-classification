```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.model_selection import GridSearchCV, RandomizedSearchCV
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.svm import SVC
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import classification_report, confusion_matrix, accuracy_score
```

```python
data = pd.read_csv(r'D:\Teks 2024\heart_classification.csv')
data.head()
```

```python
data = pd.get_dummies(data, drop_first=True)
data.head()
```

```python
x = data.drop(columns = ['HeartDisease'])
y = data['HeartDisease']
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, stratify=y, random_state=1)
```

```python
scaler = StandardScaler()
x_train = scaler.fit_transform(x_train)
x_test = scaler.transform(x_test)
```

```python
log_reg = LogisticRegression()
dec_tree = DecisionTreeClassifier()
rand_forest = RandomForestClassifier()
svm = SVC()
knn = KNeighborsClassifier()
```

```python
def evaluate_model(model, x_train, x_test, y_train, y_test):
    model.fit(x_train, y_train)
    y_pred = model.predict(x_test)
    print(f"\n{model.__class__.__name__} Results:")
    print(confusion_matrix(y_test, y_pred))
    print(classification_report(y_test, y_pred))
    print(f"Accuracy: {accuracy_score(y_test, y_pred)}")
```

```python
model = DecisionTreeClassifier(criterion = 'entropy', max_depth=10, min_samples_split = 8)
model.fit(x_train, y_train)
y_pred = model.predict(x_test)
print(accuracy_score(y_test, y_pred))
```

```python
grid = {
    'criterion' : ['gini', 'entropy'],
    'max_depth' : [6,8,10,12],
    'min_samples_split' : [3,4,5,6]
}
```

```python
gscv = GridSearchCV(estimator=model, cv=5, n_jobs=-1,param_grid=grid)
gscv.fit(X=x, y=y)
```
